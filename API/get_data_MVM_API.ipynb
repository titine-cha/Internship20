```python
# Script that remotely gets data from MVM servers
# Authorship Fantine Chabernaud

import json
import requests
import csv 
import pandas as pd

def get_site_id_list():
#Function that gets the list of site id of lake stations
#It requires the list as csv format, downloaded from MVM website

    #load dataframe from csv
    df = pd.read_csv("sj√∂_station_id_list.csv")  
    site_list = pd.to_numeric(df.columns.values.tolist())
    return site_list

def api_request_data():
    #Remove the '#' the line below before running the script /!\/!\/!\
    site_list = get_site_id_list()

    for n in site_list:
        params = {"token": "R398SAJDKLKA9AKJ2", "siteid": n, "fromYear": 1960, "toYear": 2020}

        response = requests.get(f"http://miljodata.slu.se/mvm/ws/ObservationsService.svc/rest/GetSamplesBySite", params=params)
        print(n, response.json())

        #Save each JSON file for each site into a folder
        title = "data_to_convert_input/mvm_dataset_site_" + str(n)+".json"
        with open(title, "w", encoding='utf-8') as json_file:
            json.dump(response.json(), json_file)

            
            
if __name__=="__main__":
    api_request_data() 
```
